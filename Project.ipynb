```python
from google.colab import auth
auth.authenticate_user()
import pandas as pd


import gspread
from google.auth import default
creds, _ = default()

gs = gspread.authorize(creds)
#sh = gs.create('Menu')


worksheet = gs.open('Menu').sheet1

cell = worksheet.cell(1, 2)

cell_value = cell.value
#print(cell_value)
print('Welcome To Our Restaurant')


def add(id, price, name):
    worksheet.append_row([id, price, name])
    print("Item added successfully.")


while True:
  print('-'*20)
  print('  üòçFood loversü§§')
  print('-'*20)
  print('1. Show Menu')
  print('2. Order food')
  print('3. Update Menu')
  print('4. Exit')
  print('-'*20)

  choice = int(input())
  if choice == 1:
    print('Our Menu')
    print('-'*20)
    rows = worksheet.get_all_values()
    df = pd.DataFrame(rows)
    print(df)


  elif choice == 2:
    print('Order food')
    print('-'*20)

    menu_dict = {int(row[0]): float(row[1]) for row in rows[1:]}
    order = {}
    order_input = input("Enter item IDs and quantities ('1:2, 2:1, 3:3') or type 'Done' to finish: ")
    if order_input == 'Done':
        print("No items were ordered.")
    else:
        items = order_input.split(', ')
        for item in items:
            item_id, quantity = item.split(':')
            item_id = int(item_id)
            quantity = int(quantity)
            if item_id not in menu_dict:
                print(f"Invalid item id {item_id}. Skipping...")
                continue
            order[item_id] = quantity

        total_bill = sum(menu_dict[item_id] * quantity for item_id, quantity in order.items())
        print("Your order:")
        for item_id, quantity in order.items():
            print(f"{quantity}x {df.iloc[item_id][2]} - TK{menu_dict[item_id] * quantity}")
        print('-'*20)
        print(f"Total bill: TK{total_bill}")
        print('Thank Youüòä')
        break


  elif choice == 3:
    print('Update menu')
    print('-'*20)
    print('1. clear')
    print('2. add')
    option = int(input())
    if option == 1:
      range_str = input("Enter the range of cells ('A1:B5'): ")

      cell_list = worksheet.range(range_str)
      for cell in cell_list:
        cell.value = ""
      worksheet.update_cells(cell_list)
      rows = worksheet.get_all_values()
      df = pd.DataFrame(rows)
      print(df)
      print('-'*20)
      print('Thank Youüòä')
      break

    elif option == 2:
      id = int(input("Enter item id: "))
      price = float(input("Enter item price: "))
      name = input("Enter item name: ")
      add(id=id, price=price, name=name)
      rows = worksheet.get_all_values()
      df = pd.DataFrame(rows)
      print(df)
      print('-'*20)
      print('Thank Youüòä')
      break


  elif choice == 4:
    print('Thank Youüòä')
    break
  else:
    print('Invalid choice, please try again')
```
